```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

crime_file_path = 'https://raw.githubusercontent.com/garywhi/CrimeIncidents/master/Crime_Incidents.csv'
crime_data = pd.read_csv(crime_file_path, index_col=0)
crime_data.columns
```

```python
crime_data.head()
```

```python
crime_data_features = ['BlockAddress', 'Date']
X = crime_data[crime_data_features]
y = crime_data.Type
X.head()
```

```python
df = crime_data[crime_data.BlockAddress.isin(crime_data.BlockAddress.value_counts().head(5).index)]
figure_size = (15, 5)
fig, ax = plt.subplots(figsize=figure_size)
sns.countplot(ax=ax, x='BlockAddress', data=df)
plt.show()
```

```python
top5_blocks = ['1400 BLOCK ALA MOANA', '700 BLOCK KEEAUMOKU ST', '2400 BLOCK KALAKAUA AVE', '2200 BLOCK KALAKAUA AVE', '981000 BLOCK MOANALUA RD']
df_top5 = crime_data[crime_data.BlockAddress.isin(top5_blocks)]
df_top5.head()
```
